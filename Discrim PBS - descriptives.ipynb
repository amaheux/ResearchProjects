##Correlations

```r

dat %>% select(PTMR_fullscale_mean_T1, PTMR_fullscale_mean_T2, PTMR_fullscale_mean_T3,
               EDS_mean_T1, EDS_Reason_ancestry_T1, EDS_Reason_gender_T1, EDS_Reason_race_T1,
                EDS_Reason_sexorient_T1, EDS_Reason_SES_T1, EDS_Reason_disability_T1,
               EDS_Reason_skincolor_T1) %>% na.omit() ->cordata
#install.packages("ggcorrplot")
library(ggcorrplot)
#first get all correlations among all variables in the dataset then round those correlations to 4 pts after decimal
corr <- round(cor(cordata), 4)
#plot correlations (google ggcorrplot for lots of fun ways to present this data)
ggcorrplot(corr, outline.col = "white", type = "lower", lab=T)
```
